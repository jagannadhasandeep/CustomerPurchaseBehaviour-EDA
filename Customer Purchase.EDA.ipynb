```python
plt.figsize = (10,20)
sns.histplot(data ,y="Age",x="AnnualIncome")
```

```python
plt.figsize = (100,100)
sns.barplot(data , y="ProductCategory", x="PurchaseStatus")
```

```python
sns.lmplot(y="NumberOfPurchases", x="TimeSpentOnWebsite", data=data)
```

```python
sns.relplot(data=data, y="DiscountsAvailed", x="LoyaltyProgram", kind="line")
```

```python
plt.figsize = (1000,1000)
sns.catplot(data=data, x="Gender", y="NumberOfPurchases")
```

```python
top_5_Annual_Income= data.nlargest(5, 'AnnualIncome')
print("\nTop 5 Annual Income:")
print(top_5_Annual_Income)
```

```python
sns.set(style="whitegrid")

plt.figure(figsize=(10, 6))
sns.histplot(data=data, x='DiscountsAvailed', hue='NumberOfPurchases', bins=10, kde=True, multiple="stack")

# Add labels
plt.xlabel('DiscountsAvailed')
plt.ylabel('NumberOfPurchases')

# Show the plot
plt.show()
```

```python
corr_matrix = data.select_dtypes(include=['number']).corr()

# Create a heatmap
plt.figure(figsize=(10, 10))
sns.heatmap(corr_matrix, annot=True, cmap=sns.cubehelix_palette(as_cmap=True), linewidths=0.5)

plt.show()
```

```python
sns.violinplot(x="Gender", y="AnnualIncome", data=data)

plt.xlabel("Gender")
plt.ylabel("AnnualIncome")
plt.show()
```

```python
sns.pairplot(data.select_dtypes(include=['number']))

plt.show()
```

```python
status_counts = data['NumberOfPurchases'].value_counts()
plt.figure(figsize=(10, 10))
plt.pie(status_counts, labels=status_counts.index, autopct='%1.1f%%', startangle=140)
plt.show()
```
